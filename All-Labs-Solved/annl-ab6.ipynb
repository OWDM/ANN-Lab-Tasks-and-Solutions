```python
import numpy as np

def logsigmoid(x):
  return 1 / (1 + np.exp(-x))

def purelin(x):
  return np.maximum(0, x)

# Define the weights and biases
W1 = np.array([[0.5, -0.3, 0.2], [0.4, 0.1, -0.5]])
b1 = np.array([[0.3], [0.2]])
W2 = np.array([[0.7, -0.4], [-0.2, 0.6]])
b2 = np.array([[0.1]])

# Set the input values
p_values = [-2, -1, 0, 1, 2]

# Compute a2 for each value of p
a2_values = []
for p in p_values:
    # Compute a1
    a1 = logsigmoid(np.dot(W1, p) + b1)

    # Compute a2
    a2 = purelin(np.dot(W2, a1) + b2)
    a2_values.append(a2)

# Print the values of p and a2
for p, a2 in zip(p_values, a2_values):
    print(f"p = {p}, a2 = {a2}")
```
